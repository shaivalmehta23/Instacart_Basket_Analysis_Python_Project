# 01 Creating the “price_label” and “busiest_day” columns

# If Statement to find the range of product

```python
# Define the function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Price range of products
df['price_range'].value_counts(dropna = False)
```

```python
# Checking the most expensive product
df['prices'].max()
```

# If-Statements with the loc() function to find the range of product

```python
# Apply the function
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'prices_range_loc'] = 'High-range product'
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'prices_range_loc'] = 'Mid-range product'
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'prices_range_loc'] = 'Low-range product'
```

### If-statement with for-loop for orders day of week

```python
result = []

for value in ords_prods_merge["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")    
```

# 02 Changing the busiest day to busiest days of the week

```python
# Removing the busiet_day
ords_prods_merge = ords_prods_merge.drop(['busiest_day'], axis = 1)
```

```python
# Creating column for busiest_days
result = []
for value in ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result.append("Busiest days")
    elif value == 3 or value == 4:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")
```

```python
# New column created with new heading
ords_prods_merge['busiest_days'] = result
```

# 03 Checking the accuracy of new column

```python
# Busiest day of the week
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# checking the results
ords_prods_merge.head()
```

##### The observation indicates that the number of Busiest and Slowest days has nearly doubled, while Regular days have decreased by almost half compared to the previous count of 'Busiest day' when the one bus day was used. Additionally, the Busiest days now have a number of orders comparable to Regular days, with only a 10% difference in the number of orders.

# 04 Creating a new column called “busiest_period_of_day”

```python
# frequency of order_hour_of_day
ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
# Create a 'for loop' code for 'Most orders', 'Fewest orders' and 'Average orders'

busiest_result = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [9, 10, 11, 12, 13, 14, 15, 16]:
        busiest_result.append("Most orders")
    elif value in [0, 1, 2, 3, 4, 5, 6, 23]:
        busiest_result.append("Fewest orders")
    else:
        busiest_result.append("Average orders")
```

```python
ords_prods_merge['busiest_period_of_day'] = busiest_result
```

# 05 Print frequency

```python
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

##### The observation indicates that the Most busiest hours for orders are from 9 to 16 and least busiest hours for orders are from 0 to 6 and 23, whereas the remaining time of the day has average no of orders.
